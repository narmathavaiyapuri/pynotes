```python
# Add and update values
user = {"name": "Narmatha", "age": 21}

user["age"] = 22
user["city"] = "Chennai"

print(user)
```

```python
# Loop through dictionary
marks = {"math": 80, "science": 90}

for subject, mark in marks.items():
    print(subject, mark)
```
