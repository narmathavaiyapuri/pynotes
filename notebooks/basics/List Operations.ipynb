```python
# Add and remove elements
numbers = [1, 2, 3]

numbers.append(4)
numbers.remove(2)

print(numbers)
```

```python
# List slicing
nums = [10, 20, 30, 40, 50]

print(nums[1:4])
print(nums[:3])
```

```python
# Loop and sum
marks = [80, 90, 70]

total = 0

for m in marks:
    total += m

print("Total:", total)
```
