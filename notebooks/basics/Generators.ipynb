```python
# Basic generator
def my_gen():
    yield 1
    yield 2
    yield 3

for val in my_gen():
    print(val)
```

```python
# Generator with loop
def count(n):
    for i in range(n):
        yield i

for num in count(5):
    print(num)
```

```python
# Using next()
def even_numbers():
    yield 2
    yield 4
    yield 6

g = even_numbers()

print(next(g))
print(next(g))
print(next(g))
```
