```python
# Basic try-except
try:
    num = int("abc")
except:
    print("Error occurred")
```

```python
# Specific exception
try:
    a = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# finally block
try:
    x = int("100")
    print(x)
except:
    print("Error")
finally:
    print("Done execution")
```
