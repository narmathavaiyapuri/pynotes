```python
#Number variable
a = 10
b = 5
sum_value = a + b

print("A value:", a)
print("B value:", b)
print("Sum:", sum_value)
```

```python
# String variable
name = "Raj"
greeting = "Hello " + name

print("Name:", name)
print("Greeting:", greeting)
```

```python
# Float variable
price = 50.5
tax = 5.5
total = price + tax

print("Price:", price)
print("Tax:", tax)
print("Total:", total)
```

```python
#Boolean variable
is_logged_in = True

if is_logged_in:
    print("User is logged in")
else:
    print("User is not logged in")
```
