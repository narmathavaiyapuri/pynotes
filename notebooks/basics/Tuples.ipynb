```python
#Basic tuple
numbers = (1, 2, 3)

print("Tuple:", numbers)
print("First element:", numbers[0])
```

```python
#Tuple with different data types
data = (10, "hello", 3.5)

print("Tuple:", data)
print("Second element:", data[1])
```

```python
#Loop through tuple
marks = (80, 90, 75)

for m in marks:
    print("Mark:", m)
```
