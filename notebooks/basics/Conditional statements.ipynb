```python
# Basic if
age = 18

if age >= 18:
    print("Eligible to vote")
```

```python
# if else
num = 7

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# if elif else
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
```

```python
# Nested if
num = 10

if num > 0:
    if num % 2 == 0:
        print("Positive Even")
```
