```python
# Write to file
file = open("test.txt", "w")

file.write("Hello Narmatha")
file.close()
```

```python
# Read from file
file = open("test.txt", "r")

data = file.read()
print(data)

file.close()
```

```python
# Append to file
file = open("test.txt", "a")

file.write("\nNew Line Added")

file.close()
```
