```python
# Loop through tuple
nums = (5, 10, 15)

total = 0

for n in nums:
    total += n

print("Total:", total)
```

```python
# Union and intersection
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)   # union
print(a & b)   # intersection
```

```python
# Difference and symmetric difference
x = {10, 20, 30}
y = {20, 40}

print(x - y)   # difference
print(x ^ y)   # symmetric difference
```
