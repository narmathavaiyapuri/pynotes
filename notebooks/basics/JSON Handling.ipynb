```python
# Convert dict to JSON
import json

data = {"name": "Narmatha", "age": 21}

json_data = json.dumps(data)

print(json_data)
```

```python
# Convert JSON to dict
import json

json_str = '{"name": "Narmatha", "age": 21}'

data = json.loads(json_str)

print(data["name"])
```
