```python
# Basic list comprehension
nums = [1, 2, 3, 4]

result = [x * 2 for x in nums]

print(result)
```

```python
# With condition
nums = [1, 2, 3, 4, 5]

result = [x for x in nums if x % 2 == 0]

print(result)
```

```python
# String list comprehension
words = ["hi", "hello", "python"]

result = [w.upper() for w in words]

print(result)
```
