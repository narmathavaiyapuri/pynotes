```python
# Basic iterator
nums = [1, 2, 3]

it = iter(nums)

print(next(it))
print(next(it))
print(next(it))
```

```python
# Loop using iterator
data = [10, 20, 30]

it = iter(data)

for val in it:
    print(val)
```

```python
# Custom iterator
class Count:
    def __init__(self, max):
        self.max = max
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= self.max:
            val = self.current
            self.current += 1
            return val
        else:
            raise StopIteration

c = Count(5)

for num in c:
    print(num)
```
